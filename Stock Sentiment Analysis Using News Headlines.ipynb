```python
import pandas as pd
```

```python
df = pd.read_csv("C:/Users/Ashish/Desktop/Python Projects/Stock Market Analysis Using News Headlines/Data.txt", encoding = "ISO-8859-1")
pd.set_option('display.max_columns', None)
```

```python
df.head()
```

```python
# Train-Test Split based on dates

train = df[df["Date"] < "20150101"]
test = df[df["Date"] > "20141231"]
```

- As this is a Time Series Data, The older date data needs to be used as train dataset and new date data needs to be used as test dataset. This is what we have done above.
- The Label Feature is the Dependent Feature 
- Remaining Features are Independent Features

```python
# Removing Special Characters
data = train.iloc[:, 2:27]
data.replace("[^a-zA-Z]", " ", regex = True, inplace = True)
```

```python
# Removing Column Names For Ease of Access

list1 = [i for i in range(25)]
new_index = [str(i) for i in list1]
data.columns = new_index
data.head(5)
```

```python
# Converting the headlines in lower-case

for index in new_index:
    data[index]=data[index].str.lower()
data.head(2)
```

```python
# combining all the columns together for the first row

" ".join(str(x) for x in data.iloc[1, 0:25])
```

```python
# Combining all the columns together for all the rows

headlines = []
for row in range(0, len(data.index)):
    headlines.append(" ".join(str(x) for x in data.iloc[row, 0:25]))
```

```python
headlines[1]
```

```python
# remove multiple spaces
headlines = pd.DataFrame(headlines)
```

```python
headlines.head()
```

```python
headlines[0] = headlines[0].apply(lambda x : " ".join(x.split()))
```

```python
df["Label"].value_counts()
```

- The Dataset is Balanced

```python
# Count Vectorizer

from sklearn.feature_extraction.text import CountVectorizer
from sklearn.ensemble import RandomForestClassifier
```

```python
## Implement Bag of Words

countvector = CountVectorizer(ngram_range=(2,2)) #n_gram is set to 2,2 means bigram is used
traindataset = countvector.fit_transform(headlines[0])
```

```python
# implement RandomForest Classifier
randomclassifier=RandomForestClassifier(n_estimators=200,criterion='entropy')
randomclassifier.fit(traindataset,train['Label'])
```

```python
# Predict for the test dataset
test_transform= []
for row in range(0,len(test.index)):
    test_transform.append(' '.join(str(x) for x in test.iloc[row,2:27]))
test_dataset = countvector.transform(test_transform)
predictions = randomclassifier.predict(test_dataset)
```

```python
# Import library to check accuracy

from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
```

```python
matrix = confusion_matrix(test["Label"], predictions)
print(matrix)
accuracy = accuracy_score(test["Label"], predictions)
print(accuracy)
report = classification_report(test["Label"], predictions)
print(report)
```

### TF-IDF Vectorizer

```python
# TF-IDF

from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.ensemble import RandomForestClassifier
```

```python
## Implement Bag of Words

tfidf = TfidfVectorizer(ngram_range=(2,2)) #n_gram is set to 2,2 means bigram is used
traindataset = tfidf.fit_transform(headlines[0])
```

```python
# implement RandomForest Classifier
randomclassifier=RandomForestClassifier(n_estimators=200,criterion='entropy')
randomclassifier.fit(traindataset,train['Label'])
```

```python
# Predict for the test dataset
test_transform= []
for row in range(0,len(test.index)):
    test_transform.append(' '.join(str(x) for x in test.iloc[row,2:27]))
test_dataset = tfidf.transform(test_transform)
predictions = randomclassifier.predict(test_dataset)
```

```python
# Import library to check accuracy

from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
```

```python
matrix = confusion_matrix(test["Label"], predictions)
print(matrix)
accuracy = accuracy_score(test["Label"], predictions)
print(accuracy)
report = classification_report(test["Label"], predictions)
print(report)
```
